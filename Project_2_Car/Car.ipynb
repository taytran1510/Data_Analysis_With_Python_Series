### 1: Data Cleaning:
* Find all Null value in the dataset. If there is any null value in any column, then fill it with the mean of that column

```python
car.isnull().sum()
```

```python
car['Cylinders'].fillna(car['Cylinders'].mean(), inplace = True)
```

```python
car.isnull().sum()
```

```python
car = car.dropna()
```
